```python
import random
time_ord = []
time_cont = []
prior = []
dic = {}
s = []
counter_1 = 0
n = 4
while True:
    try: 
#         choice = int(input('Введите 0 - невытесняющий, 1 - вытесняющий: '))
#         assert choice == 0 or choice == 1, 'Вы сделали неправильный выбор :('

        for i in range(n):
            s.append(random.randint(0, 5))
            s.append(random.randint(1, 5))
            s.append(random.randint(1, 5))
            assert s[0] >= 0, 'Время в очереди < 0'
            time_ord.append(s[0])
            assert s[1] > 0, 'Время выполнения <= 0'
            time_cont.append(s[1])
            assert s[2] > 0, 'Приоритет <= 0'
            prior.append(s[2])
            print(f"p{i} {prior[i]} {time_ord[i]} {time_cont[i]}")
            dic[i] = []
            s = []
        print()
            
        a = []
        k = 0
        counter = 0
        while sum(time_cont)!=0:
            minn = 100000000000
            min_ind = -1
            min_prior = 100000
            for i in range(n):
                if prior[i] < min_prior:
                    if time_ord[i] <= k and time_cont[i] != 0:
                        minn = time_cont[i]
                        min_ind = i
                        min_prior = prior[i]
                elif prior[i] == min_prior:
                    if time_cont[i] < minn and time_ord[i] <= k and time_cont[i] != 0:
                        minn = time_cont[i]
                        min_ind = i
                        min_prior = prior[i]
            if min_ind != -1:
                dic[min_ind].append('И')
                time_cont[min_ind] -= 1
                counter_1 += 1
            for j in range(n):
                if j != min_ind and time_ord[j] <= k and time_cont[j] != 0:
                    dic[j].append('Г')
                    counter += 1
                    counter_1 += 1
                elif time_cont[j] != 0 and time_ord[j] > k:
                    dic[j].append('-')
            k+=1
        for i in range(n):
            print(f"p{i}", end=' ')
            for j in dic[i]:
                print(j, end=' ')
            print()

        print(f'Среднее время ожидания равно {counter/n}')
        print(f'Среднее полное время выполнения равно {counter_1/n}')
        break
    except AssertionError as e:
        print(e)
    except KeyboardInterrupt:
        print('Программа выдержала!')
        break
    except Exception as e:
        print('Что-то намудрили')
        print(e)
```
